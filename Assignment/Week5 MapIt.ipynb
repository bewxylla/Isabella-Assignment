```python
import matplotlib.pyplot as plt

roses = ['Red', 'Yellow', 'Pink', 'Black', 'Orange']
count = [2, 3, 1, 2, 1]

plt.bar(roses, count, color=['#CE3D1E', '#FFD700', '#EA42A2', '#080507', '#FF8F1E'])

plt.title('My Rose Collection')
plt.xlabel('Rose Type')
plt.ylabel('Number of Roses')

plt.show()
```

```python
import matplotlib.pyplot as plt

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
water_intake = [8, 10, 7, 6, 11, 9, 12]

plt.plot(days, water_intake)

plt.title('Daily Water Intake')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Glasses')

plt.show()
```

```python
import matplotlib.pyplot as plt

screen_time = {
    'Phone': 53,
    'Laptop': 24,
    'TV': 12,
    'Tablet': 5,
    'Other': 3
}

plt.pie(screen_time.values(), labels=screen_time.keys(), autopct='%1.1f%%')

plt.title('Daily Screen Time')
plt.axis('equal')

plt.show()
```
